```python
def identify_outliers(df,column):
    # Step 1: Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df_rice_analysis['rice_area_1000_ha'].quantile(0.25)
    Q3 = df_rice_analysis['rice_area_1000_ha'].quantile(0.75)
    
    # Step 2: Calculate the IQR
    IQR = Q3 - Q1
    
    # Step 3: Define the bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Step 4: Identify the outlier rows
    outliers = df[(df_rice_analysis['rice_area_1000_ha'] < lower_bound) | (df_rice_analysis['rice_area_1000_ha'] > upper_bound)]
    return outliers, lower_bound, upper_bound
```

```python
#Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)


wheat_df=df.groupby('state_name')['wheat_production_1000_tons'].sum().reset_index()
top5_df=wheat_df.sort_values(by='wheat_production_1000_tons',ascending=False).head(5)

# Calculate Percentages for the Pie Chart
total_production = top5_df['wheat_production_1000_tons'].sum()
top5_df['Percentage'] = (top5_df['wheat_production_1000_tons'] / total_production) * 100


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

#BAR CHART (Top 5 States)
colors = ['skyblue', 'turquoise', 'lawngreen', 'lightcoral', 'blueviolet']
ax1.bar(top5_df['state_name'], top5_df['wheat_production_1000_tons'], color=colors, edgecolor='black', alpha=0.8)
ax1.set_title('Top 5 Wheat Producing States', fontsize=14, fontweight='bold')
ax1.set_ylabel('Production (1000 Tons)', fontsize=12)
ax1.set_xlabel('States', fontsize=12)
ax1.grid(axis='y', linestyle='--', alpha=0.5)

# Adding value labels on top of bars
for i, v in enumerate(top5_df['wheat_production_1000_tons']):
    ax1.text(i, v + 0.5, str(v), ha='center', fontweight='bold')

# --- PIE CHART 
explode = (0.1, 0, 0, 0,0)  # Explode the 1st slice (UP)
ax2.pie(top5_df['Percentage'], 
        labels=top5_df['state_name'],
        autopct='%1.1f%%', 
        startangle=140, 
        colors=plt.cm.Paired.colors, 
        explode=explode, 
        shadow=True)
ax2.set_title('Share of Total Wheat Production (%)', fontsize=14, fontweight='bold')

# 3. Final Adjustments
plt.suptitle('India Wheat Production Analysis: State-wise Breakdown', fontsize=18, y=1.02)
plt.tight_layout()
plt.show()
```

```python
#Impact of Area Cultivated on Production (Rice, Wheat, Maize)


state_summary = df.groupby(['state_name']).agg({
    'rice_production_1000_tons': 'sum',
    'rice_area_1000_ha': 'sum',
    'wheat_area_1000_ha': 'sum',
    'wheat_production_1000_tons': 'sum',
    'maize_area_1000_ha': 'sum',
    'maize_production_1000_tons': 'sum'
}).reset_index()

# --- 2.Yield Calculation ---
state_summary['Rice_Yield'] = state_summary['rice_production_1000_tons'] / state_summary['rice_area_1000_ha']
state_summary['Wheat_Yield'] = state_summary['wheat_production_1000_tons'] / state_summary['wheat_area_1000_ha']
state_summary['Maize_Yield'] = state_summary['maize_production_1000_tons'] / state_summary['maize_area_1000_ha']

# --- 3.Correlation Analysis---
# We calculate correlation across all states to see the impact of area on production
rice_corr = state_summary['rice_area_1000_ha'].corr(state_summary['rice_production_1000_tons'])
wheat_corr = state_summary['wheat_area_1000_ha'].corr(state_summary['wheat_production_1000_tons'])
maize_corr = state_summary['maize_area_1000_ha'].corr(state_summary['maize_production_1000_tons'])

print(f"Correlation (Area vs Production):")
print(f"Rice: {rice_corr:.4f}")
print(f"Wheat: {wheat_corr:.4f}")
print(f"Maize: {maize_corr:.4f}")

# --- 4. Visualizations ---
crops = [
    ('rice_area_1000_ha', 'rice_production_1000_tons', 'Rice'),
    ('wheat_area_1000_ha', 'wheat_production_1000_tons', 'Wheat'),
    ('maize_area_1000_ha', 'maize_production_1000_tons', 'Maize')
]

for area_col, prod_col, name in crops:
    sns.lmplot(x=area_col, y=prod_col, data=state_summary, height=6, aspect=1.5)
    plt.title(f'Overall Impact of {name} Cultivated Area on Production')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

```
